```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
my_pca = PCA(n_components=30)
pos_emb_pca = my_pca.fit_transform(pos_emb) 
```

```python
pos_emb_pca.shape
```

```python
pos_emb_pca[:,0].shape
```

```python
from numpy import dot
from numpy.linalg import norm
def cos_sim(a,b):
    return dot(a, b)/(norm(a)*norm(b))
```

```python
cs_pca = [cos_sim(pos_emb_pca[0,:], pos_emb_pca[i,:]) for i in range(5,258)]
```

```python
cs = [cos_sim(pos_emb[0,:], pos_emb[i,:]) for i in range(1,258)]
```

```python
plt.plot(range(1,258), cs)
```

```python
plt.plot(range(5,258), cs_pca)
```
